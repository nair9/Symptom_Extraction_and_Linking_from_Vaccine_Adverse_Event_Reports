**Calculating Similarity between Ground Truth and Predictions**

```python
import torch
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity

tokenizer = BertTokenizer.from_pretrained('dmis-lab/biobert-base-cased-v1.2')
model = BertModel.from_pretrained('dmis-lab/biobert-base-cased-v1.2')
```

```python
y_true = ground_truth
y_pred = test_df['Standard_Linked']
```

```python
res = dict()
for true, pred in tqdm(zip(y_true,y_pred),total=100):
    true_tokens = tokenizer(true, return_tensors='pt')
    pred_tokens = tokenizer(pred, return_tensors='pt')

    with torch.no_grad():
        true_emb = model(**true_tokens).last_hidden_state.mean(dim=1)
        pred_emb = model(**pred_tokens).last_hidden_state.mean(dim=1)

    true_emb = true_emb.detach().numpy()
    pred_emb = pred_emb.detach().numpy()

    res[(pred,true)] = cosine_similarity([np.ravel(true_emb)],[np.ravel(pred_emb)])[0][0]

# max_key, max_val = max(scores.items(), key= lambda x: x[1])
# similarity_scores[symp] = [max_key, max_val]
#     break
```

**Average of Similarity Scores between Ground Truth and Predictions**

```python
res_avg_score = sum(res.values())/len(res.values())
```

```python
print(f"Average Similarity Score between Ground Truth and Predictions: {res_avg_score:.2f}")
```

**Getting similarity scores for extracted words with the ground truth and using this to compare the similarity scores with linked symptoms**

```python
true_res = dict()
for true, pred in tqdm(zip(y_true,test_symptoms),total=100):
    true_tokens = tokenizer(true, return_tensors='pt')
    pred_tokens = tokenizer(pred, return_tensors='pt')

    with torch.no_grad():
        true_emb = model(**true_tokens).last_hidden_state.mean(dim=1)
        pred_emb = model(**pred_tokens).last_hidden_state.mean(dim=1)

    true_emb = true_emb.detach().numpy()
    pred_emb = pred_emb.detach().numpy()

    true_res[(pred,true)] = cosine_similarity([np.ravel(true_emb)],[np.ravel(pred_emb)])[0][0]
```
