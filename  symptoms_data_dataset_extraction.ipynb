```python
import os
import pandas as pd
```

```python
vax_data = pd.read_csv("SAMPLED_DATA.csv")
```

```python
symptoms_data = pd.DataFrame()
```

```python
path = "dataset/"
```

```python
for i in os.listdir(path):
#     print(i)
    if "SYMPTOMS" in i:
        data = pd.read_csv(os.path.join(path,i), encoding='ISO-8859-1')
        symptoms_data = pd.concat([symptoms_data,data], axis=0, ignore_index=True)
```

```python
vaers_id = vax_data["VAERS_ID"].unique()
```

```python
vax_data["VAERS_ID"].nunique()
```

```python
vax_data["VAERS_ID"].count()
```

```python
vax_symptoms_data = pd.DataFrame()
for i in vaers_id:
    if int(i) in symptoms_data["VAERS_ID"]:
        vax_symptoms_data = pd.concat([vax_symptoms_data,symptoms_data[symptoms_data["VAERS_ID"]==int(i)]], axis = 0, ignore_index= True)
```
